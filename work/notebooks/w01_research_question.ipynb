# 1. My Lane

**Refresh / Content Opportunity Scoring**

## Why I Chose This Lane

I chose the **Refresh / Content Opportunity Scoring** lane because it addresses a practical business problem that many organizations face: deciding which content should be reviewed first when time and resources are limited.

Large websites often contain thousands of pages, making it unrealistic for editors to manually evaluate every page. Instead of attempting to predict whether a page will become successful, this project focuses on helping content teams prioritize which pages deserve human attention based on observable search and engagement signals.

This lane also provides an opportunity to develop a complete machine learning workflow throughout the internship. It combines exploratory data analysis, feature engineering, baseline scoring, supervised learning, model evaluation, and explainable recommendations. The final objective is not to replace human judgment, but to support better and more consistent decision-making.

## 2. The question: decision, action, cost of a wrong call

---
## The Question

**Which content pages should be prioritized for review based on their search and engagement signals?**

## Decision

Given limited editorial capacity, which pages should the content team review first?

---

## Action

Editors can review the highest-ranked pages first and decide whether improvements are necessary.

---

## Cost of a Wrong Call

A false positive means editors spend valuable time reviewing pages that were already performing well or did not require immediate attention.

A false negative means genuinely important pages may continue losing visibility or engagement because they were not identified early enough.

Both types of mistakes reduce the effectiveness of limited editorial resources.



```python
# Cloning my repository
!git clone https://github.com/innouguru/flyrank_ml_track_1.git
```

```python
import pandas as pd

df = pd.read_csv(
    "flyrank_ml_track_1/data/raw/content_refresh_anonymized.csv"
)

df.head()
```

```python
total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
declining_pct = declining_pages / total_pages * 100

avg_impressions = df["impressions_90d"].mean()
avg_age = df["content_age_days"].mean()

print(f"Total pages: {total_pages:,}")
print(f"Declining pages: {declining_pages:,} ({declining_pct:.1f}%)")
print(f"Average impressions (90d): {avg_impressions:,.0f}")
print(f"Average content age: {avg_age:.0f} days")
```
