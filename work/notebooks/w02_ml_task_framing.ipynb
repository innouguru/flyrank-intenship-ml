## 3. Success metric

**Success Metric: Precision@K.**

Precision@K measures the proportion of the top K pages recommended by the model that are truly good candidates for review. This metric is appropriate because the content team has limited time and focuses on reviewing only the highest-priority pages rather than every page that may need attention. Manual review of the top recommendations can then be used to verify that the model's suggestions are meaningful and useful for the business.

**A good result:** Precision@50 ≥ 0.80, meaning at least 40 of the top 50 recommended pages are genuinely good candidates for review. This would allow the content team to spend most of their effort on worthwhile pages instead of reviewing many low-value recommendations.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule is insufficient because it must consider many features and the complex relationships between them. The importance of one feature may depend on the values of other features, making it difficult to define a small set of rules that works for every content page. An ML model can learn these patterns from data and produce more reliable recommendations for which pages should be prioritized for review.
